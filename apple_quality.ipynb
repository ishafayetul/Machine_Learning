```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("drive/MyDrive/Learning/ML/Logistics Regression/LR_Binary Classification/Apple Quality/apple_quality.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
pdf=le.fit_transform(df.Quality)
pdf
```

```python
ndf=df.drop(["A_id","Quality"],axis="columns")
ndf.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(ndf,pdf,test_size=0.2)
```

```python
model=LogisticRegression(max_iter=500)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
