```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
hiring_df=pd.read_csv("drive/MyDrive/Learning/ML/Hiring Multi Var/hiring.csv")
hiring_df
```

```python
modified_hiring_df=hiring_df.rename(columns={"experience":"Experience","test_score(out of 10)":"Test_Score","interview_score(out of 10)":"Interview_Score","salary($)":"Salary"})
```

```python
from word2number import w2n
for x in range(2,8):
  modified_hiring_df["Experience"][x]=w2n.word_to_num(modified_hiring_df["Experience"][x])
modified_hiring_df
```

```python
#modified_hiring_df.Experience=modified_hiring_df.Experience.fillna(modified_hiring_df.Experience.median())
#modified_hiring_df.Test_Score=modified_hiring_df.Test_Score.fillna(modified_hiring_df.Test_Score.median())

modified_hiring_df.Experience=modified_hiring_df.Experience.fillna(0)
modified_hiring_df.Test_Score=modified_hiring_df.Test_Score.fillna(modified_hiring_df.Test_Score.median())
```

```python
modified_hiring_df
```

```python
model=linear_model.LinearRegression()
dropped_df=modified_hiring_df.drop("Salary",axis="columns")
model.fit(dropped_df,modified_hiring_df.Salary)
```

```python
model.predict([[12,10,10]])
```
