```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("drive/MyDrive/Learning/ML/Logistics Regression/LR_Binary Classification/HR Analytics/HR_comma_sep.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() #LE converts string into numeric value
```

```python
sal_dummy=pd.get_dummies(df.salary,prefix="salary")
sal_dummy.head()
```

```python
dept_dummy=pd.get_dummies(df.Department,prefix="dept")
dept_dummy.head()
```

```python
df=pd.concat([df,dept_dummy,sal_dummy],axis="columns").drop(["salary","Department"],axis="columns")
df.head()
```

```python
df.Department = le.fit_transform(df.Department)
df.salary = le.fit_transform(df.salary)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(df[["satisfaction_level","number_project","average_montly_hours","time_spend_company","Work_accident","promotion_last_5years","dept_IT","dept_RandD","dept_hr","dept_management","dept_marketing","dept_product_mng","dept_sales","dept_support","dept_technical","salary_high","salary_low","salary_medium"]],df.left,test_size=0.5)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
import joblib
joblib.dump(model,"drive/MyDrive/Learning/ML/Logistics Regression/LR_Binary Classification/HR Analytics/model")
```
