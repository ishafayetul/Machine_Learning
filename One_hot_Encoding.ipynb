```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
price_df=pd.read_csv("drive/MyDrive/Learning/ML/Dummy Variables & One Hot Encoding/homeprices.csv")
```

```python
price_df
```

#Dummy Variable Method

```python
dummies=pd.get_dummies(price_df)
dummies
```

```python
dummies_df=dummies.drop(["town_west windsor"],axis="columns")
```

```python
dummies_df
```

```python
x=dummies_df.drop(["price"],axis="columns")
x
```

```python
y=dummies_df.price
y
```

```python
model= linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[2000,0,1]])
```

```python
model.score(x,y)
```

#One Hot Encodding


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() #LE converts string into numeric value
```

```python
dfle=price_df
dfle.town = le.fit_transform(dfle.town) #converting town column's nominal values into numeric values
dfle
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
ohe_x=ohe.fit_transform(dfle[["town"]])
ohe_x
```

```python
join_x=np.hstack((ohe_x,dfle))
join_x
```

```python
x=np.delete(join_x,[0,3,5],1)
x
```

```python
enc_model=linear_model.LinearRegression()
enc_model.fit(x,y)
```

```python
enc_model.predict([[0,1,3400]])
```
