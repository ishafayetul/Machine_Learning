```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/drive/MyDrive/Learning/ML/Income Prediction/per_capita_income.csv")
```

```python
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(df.year,df.income)
```

```python
year_df=df.drop("income",axis="columns")
```

```python
model=linear_model.LinearRegression()
model.fit(year_df,df.income)
```

```python
model.predict([[2020]])
```
